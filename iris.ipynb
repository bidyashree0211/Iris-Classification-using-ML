**VISUALIZATION OF DATASET**

```python
#Visualize the whole dataset
sns.pairplot(cf,hue="Class_label") #different color combin ations for different class labels
```

**SEPARATING INPUT AND OUTPUT COLUMNS**

```python
data = cf.values #data in matrix form
#Creating two new matrix after slicing them
X = data[:,0:4]
Y = data[:,4]
print(X)#-->all rows except Class_Labels Row
print(Y)#-->onlyClass_Labels Row
```

**SEPARATING DATA INTO TRAINING AND TESTING**

```python
#Split the dataset into train and test datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2) #120 Rows and 30 Rows = 0.2
#print(Y_train)
#print(X_train)
#print(Y_test)
#print(X_test)
```

**MODEL 1: Support vector machine algorithm**

```python
#Support vector machine  algorithm
from sklearn.svm import SVC

model_svc = SVC() #model fitting by training the system
model_svc.fit(X_train , y_train)
```

```python
predict1 = model_svc.predict(X_test)
#Calculate accuracy of each flower
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test , predict1)*100) #comparing predictions with testing output
#for i in range(len(predict1)):
 # print(y_test[i],predict1[i])--->To check the accuracy obtained


```

**MODEL2: LOGISTIC REGRESSION**

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train,y_train)
```

```python
predict2 = model_LR.predict(X_test)
#Calculate accuracy of each flower
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test , predict2)*100)
```
